```python
image_path = 'results/spongebob.png'  
image = cv2.imread(image_path)                         # load image

if image is None:                                      # check if image loaded
    print('Error loading image')
else: 
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)     # convert to rgb
    plt.imshow(image)
    plt.axis("off")
    plt.show()
```

```python
print(image.shape)                       # image dimension - height, width, color channels
h,w = image.shape[:2]                    # extract height and width
centerX, centerY= (w // 2), (h // 2)     # center coordinates
```

```python
# w/2 does not work because its not integer (540/2 = 270.0)
# / always returns floating-point number even if result is a whole number
# // performs floor division

# slicing begins at (0,0) <- top left of image
# slicing follows same convention as .shape method
# slice[h, w]

# top left 
# -------
# height = from 0 to half way down the height
# width = from 0 to half way down the width
topLeft2 = image[0:(h//2), 0:(w//2)]  
plt.imshow(topLeft2)
plt.axis("off")
plt.show()

print(topLeft2.shape)
```

```python
# bottom left
# -------
# height = from half way down the height to end of image
# width = from 0 to half way down the width

bottomLeft2 = image[h//2:, 0:w//2]
plt.imshow(bottomLeft2)
plt.axis("off")
plt.show()
```

```python
slice = image[0:h//2, 0:w//3]
plt.imshow(slice)
plt.axis("off")
plt.show()
```
