```python
image_path = 'results/spongebob.png'  
image = cv2.imread(image_path)                         # load image

if image is None:                                      # check if image loaded
    print('Error loading image')
else: 
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)     # convert to rgb
    plt.imshow(image)
    plt.show()
```

```python
print(image.shape)                       # image dimension - height, width, color channels
h,w = image.shape[:2]                    # extract height and width
centerX, centerY= (w // 2), (h // 2)     # center coordinates
```

```python
print(image)
```

```python
# w/2 does not work because its not integer (540/2 = 270.0)
# / always returns floating-point number even if result is a whole number
# // performs floor division
topLeft2 = image[0:(w//2), 0:(h//2)]  
plt.imshow(topLeft2)
plt.axis("off")
plt.show()

print(topLeft2.shape)
```

```python
topLeft = image[0:centerY, 0:centerX]
plt.imshow(topLeft)
plt.axis("off")
plt.show()
```

```python
bottomLeft = image[centerY:h, 0:centerX]
cv2.imshow("BottomLeft", bottomLeft)
cv2.waitKey(0)
```

```python
topRight = image[0:centerY, centerX: w]
cv2.imshow("TopRight", topRight)
cv2.waitKey(0)
```

```python
bottomRight = image[centerY:h, centerX:w]
cv2.imshow("BottomRight", bottomRight)
cv2.waitKey(0)
```
